```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.models import load_model
from keras.models import Sequential
from keras.layers import Dense, Dropout, Embedding, LSTM, Bidirectional, SpatialDropout1D
from keras import optimizers
import pickle
```

Plot graph of training and validation accuracy with loss.

```python
def plot_graph(hist):
    acc = hist['acc']
    val_acc = hist['val_acc']
    loss = hist['loss']
    val_loss = hist['val_loss']

    epochs = range(1, len(acc) + 1)

    plt.plot(epochs, acc, 'bo', label='Training acc')
    plt.plot(epochs, val_acc, 'b', label='Validation acc')
    plt.title('Training and validation accuracy')
    plt.legend()

    plt.figure()

    plt.plot(epochs, loss, 'bo', label='Training loss')
    plt.plot(epochs, val_loss, 'b', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()

    plt.show()
```

<h3>Models Performance<h3>

<h4>Model 1) Learning rate - 0.01, Input length - 6000 without pre-trained embeddings</h4>

```python
with open('trained_models/0.01_6000_hist','rb') as f:
    hist = pickle.load(f)
plot_graph(hist)    
```

Observing the gap becoming wider between Training and validation loss the model is overfitting after five epochs

<h4>Model 2) Learning rate - 0.01, Input length - 8000 without pre-trained embeddings</h4>

```python
with open('trained_models/0.01_8000_hist','rb') as f:
    hist2 = pickle.load(f)
plot_graph(hist2)    
```

Model performs better than the previous one however more tuning is needed like increasing input dimensions or reducing number of hidden units

<h4>Model 3)Bidirectional LSTM Learning rate - 0.01, Input length - 1000 without pre-trained embeddings</h4>

```python
with open('trained_models/1000_bd_hist','rb') as f:
    hist3 = pickle.load(f)
plot_graph(hist3) 
```
